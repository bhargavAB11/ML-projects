```python
l=LabelEncoder()
data['diagnosis']=l.fit_transform(data['diagnosis'])
```

```python
x=data.drop(['diagnosis'],axis=1)
y=data['diagnosis']
```

# Train model

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
scale=StandardScaler()
x_train=scale.fit_transform(x_train)
x_test=scale.fit_transform(x_test)
y_train1=(y_train==0)
y_test1=(y_test==0)
```

```python
from sklearn.linear_model import LogisticRegression
log_clf=LogisticRegression(tol=0.0001,C=1.0,random_state=42,solver='lbfgs',max_iter=1000)
log_clf.fit(x_train,y_train1)
```

```python
from sklearn.model_selection import cross_val_score
cross_val_score(log_clf,x_train,y_train1,cv=3)
```

```python
from sklearn.model_selection import cross_val_predict
cross_pred=cross_val_predict(log_clf,x_train,y_train1,cv=3)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_train1,cross_pred)
```

```python
from sklearn import metrics
print(metrics.classification_report(y_train1,cross_pred))
```

```python
y_train_score=cross_val_predict(log_clf,x_train,y_train1,cv=3,method="decision_function")
```

```python
from sklearn.metrics import roc_auc_score
roc_auc_score(y_train1,cross_pred)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_train1,cross_pred)
```

```python
from sklearn.metrics import roc_curve
fpr,tpr,thresholds=roc_curve(y_train1,y_train_score)
plt.plot(fpr,tpr,lw=2,)
plt.plot([0,1],[0,1],"k--")
plt.ylabel("FPR")
plt.xlabel("TPR")
plt.grid(True)
plt.show
```
